**Data Visualization II**

1. Use the inbuilt dataset 'titanic' as used in the above problem. Plot a box plot for distribution of age with respect to each gender along with the information about whether they survived or not. (Column names: 'sex' and 'age')

2. Write observations on the inference from the above statistics.

**About the dataset:** The dataset has 891 rows and 15 columns. However, there are only 2 numeric columns, namely, Fare and Age.

```python
# Import necessary libraries
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme()

%matplotlib inline
```

```python
# Load the dataset into the dataframe
raw_data = sns.load_dataset('titanic')

raw_data.head()
```

```python
# Print the number of rows in the dataset
print(f'The number of rows in the dataset is: {raw_data.shape[0]}')
print(f'The number of columns dataset is: {raw_data.shape[1]}')
```

### Data Preprocessing

```python
# .info() is used to get summary of dataset
raw_data.info()
```

**Note: Since the problem statement states that we only have to deal with sex and age columns, only null values in those columns will be replaced.**

```python
# Check for null values in the dataset
raw_data.isna().sum()
```

```python
# Replacing null ages with the mean age 
# Point to note: Age and fare are the only Numeric columns
mean_age = raw_data['age'].mean()

print(f'The mean age of passengers is: {mean_age: .2f}')

raw_data['age'].fillna(mean_age, inplace=True)
```

### Visualization

```python
# Plotting the barplot of "Age" column wrt Sex and survival

plt.figure(figsize=(8,8))

sns.barplot(x='sex', y='age', hue='survived', data=raw_data)

plt.title('Mean Age of Passengers wrt Sex (Barplot)')
plt.legend(title='Survived', bbox_to_anchor=(1, 1), loc=2)
plt.show()
```

**Observations**

- The mean age of males who did not survive is: 30-35
- The mean age of males who survived is: 25-30
- The mean age of females who did not survive is: 25-28
- The mean age of females who survived is: 28-30

```python
# Plotting the boxplot between age and sex

plt.figure(figsize=(8,8))

sns.boxplot(x='sex', y='age', hue='survived', data=raw_data)

plt.show()
```

**Observations**

- Almost 50 Percent of the passengers where between the age of 20 and 40. This is evident from the fact that the interquartile range of the passenger age is between 20-40.

```python
# Plotting the swarmplot between sex and age

plt.figure(figsize=(8,8))

sns.swarmplot(x='sex', y='age', hue='survived', data=raw_data, size=5)

plt.show()
```

**Observations**

- The high density of markers in the middle of the plot confirm our earlier made hypothesis that most of the passengers are between 20-40 years of age.
- The dominance of the blue hue in males shows that less than 50 percent of the males did not survive the disaster.
- On the other hand, the dominance of the orange hue indicates that more than 50 percent of females survived.
- It also shows that the oldest male onboard was 80 years old (survived) and the oldest female onboard was 60-70 years old.

```python
# Plotting the violinplot 

plt.figure(figsize=(8,8))

sns.violinplot(x='sex', y='age', hue='survived', data=raw_data)

plt.show()
```

**Observations**

- The width of the density plot indicates that a large number of passengers were between the age of 20-40.

```python
# Plotting the strippplot between sex and age

plt.figure(figsize=(6,6))

sns.stripplot(x='sex', y='age', hue='survived', data=raw_data)

plt.show()
```

**Observations**

- The stripplot confirms our earlier anaylsis that more than 50 percent of the males died in the disaster, whereas more than 50 percent of the females survived.
- It also shows that the oldest male onboard was 80 years old (survived) and the oldest female onboard was 60-70 years old.

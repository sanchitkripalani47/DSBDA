```python
# Import necessary libraries 
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

%matplotlib inline
```

```python
# Load the dataset into a DataFrame

read_data = pd.read_csv("Datasets/iris.csv")
raw_data = pd.DataFrame(read_data)
```

```python
# Check for null values in the dataset 
raw_data.isna().sum()
```

```python
# Use the .unique() method to get unique values in each column

uni_sep_len = raw_data['sepal.length'].unique().size
uni_sep_width = raw_data['sepal.width'].unique().size
uni_pet_len = raw_data['petal.length'].unique().size
uni_pet_width = raw_data['petal.width'].unique().size
uni_variety = raw_data['variety'].unique().size

print(f'The number of unique values in sepal len column is: {uni_sep_len}')
print(f'The number of unique values in sepal width column is: {uni_sep_width}')
print(f'The number of unique values in petal len column is: {uni_pet_len}')
print(f'The number of unique values in petal width column is: {uni_pet_width}')
print(f'The number of unique values in variety column is: {uni_variety}')
```

**Observations**

- Therefore, it can be said Sepal Length, Sepal Width, Petal Length, Petal Width all are numeric columns.
- The column Variety is a Categorical variable as it has only 3 distinct values. It is a nominal categorical varable since there is no specfic order to the distinct values.

### Visualization

```python
# Function to plot multiple subplots
def plot_data(plot_name):
    x = ['sepal.length', 'sepal.width','petal.length', 'petal.width']

    fig, axes = plt.subplots(2,2, figsize=(12,12))
    axes_flat = axes.flatten()


    for index, col in enumerate(x):
        axis = axes_flat[index]
        if plot_name == 'hist':
            sns.histplot(data=raw_data, x=col, hue='variety', ax=axis)
            fig.suptitle("Histogram of Attributes wrt Species (Variety)")
            
        elif plot_name == 'box':
            sns.boxplot(data=raw_data, x='variety', y=col, ax=axis)
            fig.suptitle("Boxplot of Attributes wrt Species (Variety)")
    
    plt.show()
```

```python
# Plotting the Histogram for all attributes wrt Variety
plot_data('hist')
```

**Observations**

- The attributes of Iris-Versicolor and Iris-Virginicaare almost similar. They are, however, different interms of Sepal length and Sepal Width.
- The petal length and width of Iris-Setosa is smaller as compared to Iris-Versicolor and Iris-Virginica.

```python
# Plotting a boxplot for each feature
plot_data('box')
```

**Observations**

The outliers can be seen with the help of the boxplots above.

- Outliers in iris-setosa can be seen wrt sepal width, petal length, petal width.
- Outliers in iris-versicolor can be seen wrt petal length.
- Outliers in iris-virginica can be seen wrt sepal length, sepal width.
